## Import Libraries

```python
import pandas as pd                         #to import the dataset
import numpy as np                          #to perform array operations
import pickle                               #to export preprocessed data
import matplotlib.pyplot as plt             #to plot statistics
import seaborn as sns                       #to plot statistics
from nltk.tokenize import word_tokenize     #to get words
from textblob import TextBlob               #to split the context into sentences
```

## Data Load

```python
df = pd.read_pickle("answerable_records.pkl")
```

## Plotting functions

```python
def draw_plots(data, xlabel, ylabel, title):
    """Function to draw PDF and Box Plots for the data distribution that is given to it. 
    Takes the data distribution, xlabel, ylabel and title of the plot as parameters"""
    
    #drawing the canvas
    fig = plt.figure(figsize = (20,6))
    
    #defining axes
    distplot = fig.add_subplot(1,2,1)
    boxplot = fig.add_subplot(1,2,2)
    
    sns.set(color_codes=True)
    
    #plotting PDF plot
    distplot.set(title = title, xlabel = xlabel, ylabel = ylabel)
    sns.distplot(data, ax = distplot)
    
    #plotting Boxplot
    boxplot.set(title = title, ylabel = xlabel)
    sns.boxplot(data, ax = boxplot, orient = "v")
    
    plt.plot()
```

## Data Analysis
### Question Statistics
### Sentence Length

```python
#splitting the question into sentences and storing sentence length as a separate column/field
df['question_sentence_len'] = df['question'].apply(lambda x: len(TextBlob(x).sentences))
```

```python
#printing statistics of question sentence length
print("Min: ", np.array(df['question_sentence_len'].values).min())
print("Max: ", np.array(df['question_sentence_len'].values).max())
print("Mean: ", np.array(df['question_sentence_len'].values).mean())
print("Median: ", np.median(np.array(df['question_sentence_len'].values)))
print("25th percentile: ", np.percentile(np.array(df['question_sentence_len'].values), 25))
print("50th percentile: ", np.percentile(np.array(df['question_sentence_len'].values), 50))
print("75th percentile: ", np.percentile(np.array(df['question_sentence_len'].values), 75))
print("95th percentile: ", np.percentile(np.array(df['question_sentence_len'].values), 95))
print("99th percentile: ", np.percentile(np.array(df['question_sentence_len'].values), 99))
```

```python
#question sentence length data distribution
draw_plots(df['question_sentence_len'].values, "Sentence Length", "Fraction of Questions", "Sentence Length")
```

```python
combined_text = df['question'] + df['context']
```

```python
char_len = []
for i in combined_text:
    for j in word_tokenize(i):
        char_len.append(len(j))
```

```python
print("Min: ", np.array(char_len).min())
print("Max: ", np.array(char_len).max())
print("Mean: ", np.array(char_len).mean())
print("Median: ", np.median(char_len))
print("25th percentile: ", np.percentile(np.array(char_len), 25))
print("50th percentile: ", np.percentile(np.array(char_len), 50))
print("75th percentile: ", np.percentile(np.array(char_len), 75))
print("95th percentile: ", np.percentile(np.array(char_len), 95))
print("99th percentile: ", np.percentile(np.array(char_len), 99))
print("99th percentile: ", np.percentile(np.array(char_len), 99.9))
```

```python
#character length data distribution
draw_plots(char_len, "Character Length", "Fraction of Words", "Character Length")
```
